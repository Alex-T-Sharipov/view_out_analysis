```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, ttest_rel

s1_exp1_impressions_df = pd.read_csv("1stSession_EXP1 Impressions.csv")
s1_exp2_impressions_df = pd.read_csv("1stSession_EXP2 Impressions.csv")
s1_exp1_background_df = pd.read_csv("1stSession_EXP1 BackgroundEnvironment.csv")
s1_exp2_background_df = pd.read_csv("1stSession_EXP2 BackgroundEnvironment.csv")


s2_exp1_impressions_df = pd.read_csv("2ndSession_EXP1 Impressions.csv")
s2_exp2_impressions_df = pd.read_csv("2ndSession_EXP2 Impressions.csv")
s2_exp1_background_df = pd.read_csv("2ndSession_EXP1 BackgroundEnvironment.csv")
s2_exp2_background_df = pd.read_csv("2stSession_EXP2 BackgroundEnvironment.csv")
```

```python
# Compare the Numerical feature ratings between the 2 experimental sessions

columns_to_compare = ["Pleasant", "Interesting", "Engaging", "Enjoy Working"]

# perform t-test + box plots
for column in columns_to_compare:
    # independent t-test - all participants
    t_stat_independent, p_value_independent = ttest_ind(
        s1_exp1_impressions_df[column], s2_exp1_impressions_df[column]
    )
    print(
        f"Experiment 1, Independent T-test for {column}: T-statistic = {t_stat_independent}, P-value = {p_value_independent}"
    )

    plt.figure(figsize=(10, 6))
    plt.boxplot(
        [s1_exp1_impressions_df[column], s2_exp1_impressions_df[column]],
        labels=["Session 1", "Session 2"],
    )
    plt.title(f"Comparison of {column} ratings between two sessions for experiment 1")
    plt.ylabel(f"{column} Rating")
    plt.show()
```

In experiment 1, ratings do not show a statistically significant difference between session 1 and 2.

```python
# same for experiment 2
# perform t-test + box plots
for column in columns_to_compare:
    # independent t-test - all participants
    t_stat_independent_2, p_value_independent_2 = ttest_ind(
        s1_exp2_impressions_df[column], s2_exp2_impressions_df[column]
    )
    print(
        f"Experiment 2, Independent T-test for {column}: T-statistic = {t_stat_independent_2}, P-value = {p_value_independent_2}"
    )

    plt.figure(figsize=(10, 6))
    plt.boxplot(
        [s1_exp1_impressions_df[column], s2_exp2_impressions_df[column]],
        labels=["Session 1", "Session 2"],
    )
    plt.title(f"Comparison of {column} ratings between two sessions for experiment 2")
    plt.ylabel(f"{column} Rating")
    plt.show()
```

Pleasant and Enjoy Working ratings in experiment 2 do not show a statistically significant difference between session 1 and 2, however Engaging and Interesting had difference. This might imply that participants' perceptions of the scenes evolved between the sessions. The significant difference in 'Interesting' ratings, in particular, is notable and could be linked to a change in the novelty or appeal of the scenes over time or due to varying weather conditions. The absence of significant differences in 'Pleasant' and 'Enjoy Working' ratings in both experiments indicates a consistent level of satisfaction and comfort with the scenes across sessions.
